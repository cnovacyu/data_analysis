```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Read the data
college = pd.read_csv("college.csv")
college.head(5)
```

```python
# b) set index for first column
college = pd.read_csv("college.csv", index_col=0)
college.head(5)
```

```python
# ci) produce numerical summary of all variables in the data set
college.describe()
```

```python
# cii) Produce scatterplot matrix for first 10 columns of data (1st column is not numerical)
pd.plotting.scatter_matrix(college.iloc[:, 2:11], figsize=(30, 20))
```

```python
# ciii) Produce side-by-side boxplots of Outstate vs Private
# Private is not numerical - is qualitative
college.boxplot(column= 'Outstate')
```

```python
# civ) create new variable Elite, by binning Top10perc
# use summary to see how many elite universities there are
college['Elite'] = np.where(college['Top10perc'] > 50, "Yes", "No")
college['Elite'].describe()
```

```python
college.hist(column='Enroll', bins=30, grid=False)
```

```python
# cv) produce histograms with different quantitative variables
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.hist(college['Enroll'], bins=50)
ax1.title.set_text('Enroll')

ax2.hist(college['Grad.Rate'], bins=50)
ax2.title.set_text('Grad Rate')

ax3.hist(college['Books'], bins=30)
ax3.title.set_text('Books')

ax4.hist(college['Expend'], bins=100)
ax4.title.set_text('Expend')

fig.tight_layout()
```

```python
# cvi) Continue exploring the data and provide a brief summary of what you discover
college.corr()
# High out-of-state tuition correlates to high graduation rate
```

```python
# The higher the instructional expenditure spent per student,  
# the higher the graduation rate
plt.scatter(college['Outstate'],college['Grad.Rate'])
plt.xlabel('Outstate')
plt.ylabel('Grad Rate')
```

```python
# There is a positive relationship between graduation rate and students from
# top 10% of high school classes, however it is not a strong relationship (0.495)
# There is a data point with a 119% graduation rate, which is invalid
# Looking at the correlation matrix, strength of relationship of graduation 
# rate for students from top 25% is similar to top 10% students
# Schools with students from top 10% don't always have the highest graduation rates
plt.scatter(college['Top10perc'],college['Grad.Rate'])
plt.xlabel('Top 10 %')
plt.ylabel('Grad Rate')
```

```python
# As the % of faculty members with terminal degrees increases, the number
# of applications received increases, but the relationship isn't very strong
plt.scatter(college['Terminal'],college['Apps'])
plt.xlabel('Faculty Terminal Degrees')
plt.ylabel('Apps Received')
```

```python
# Colleges with high acceptance rates receive a high number of applications
plt.scatter(college['Apps'],college['Accept'])
plt.xlabel('Apps Received')
plt.ylabel('Apps Accepted')
```
