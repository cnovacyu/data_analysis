```r
# e) Use * and : to fit linear regression models with interaction effects.
# Do any interactions appear to be statistically significant?

# I chose three variable sets that had the highest correlation in the 
# correlation matrix to check if there were any interaction effects between 
# them. There appears to be a statistically significant interaction effect 
# between displacement:horsepower since the p-value is very close to 0.
# This means that the true relationship is not additive; some predictors are
# not completely independent of other predictors. P-value is high for cylinders:
# displacement and displacement:weight, meaning there is no statistically
# significant interaction effect between those variable sets.

lm.fit3=lm(mpg~displacement*cylinders+displacement*weight
           +displacement*horsepower,data=Auto)
summary(lm.fit3)
```

```r
# f) Try a few different transformations of the variables, such as log(X),
# sqrt(X), X^2. Comment on your findings

# I want to first start with refining my previous model. I removed cylinders 
# and acceleration as predictors since they had high p-values in the original
# multiple linear regression. I then also added in the interaction
# displacement:horsepower variables to see where my new baseline was. R2 is 85.8,
# F statistic is much larger than 1 with a p-value very close to 0.

lm.fit4=lm(mpg~.-name-cylinders-acceleration+displacement:horsepower, data=Auto)
summary(lm.fit4)
```

```r
# Noticeable pattern in residuals have decreased a bit, but still have a bit
# of an effect from a few high outliers. Residual vs fitted still indicates
# a non-linear realationship between some predictors and response
par(mfrow=c(2,2))
plot(lm.fit4)
```

```r
# squaring horsepower seems to have no statistical significance and increases
# the p-value. Similary using log on origin increases the p-value slightly.
# Overall, fit is still pretty similar - F statistic decreased a bit, but still
# has a very low p-value. R2 increased a bit as well
lm.fit5=lm(mpg~displacement+horsepower+I(horsepower^2)+sqrt(weight)+year+
           log(origin)+displacement:horsepower)
summary(lm.fit5)
```

```r
# Not much significant change in results from modified model. Overall 
# still a good fit

lm.fit6=lm(mpg~displacement+horsepower+weight+I(weight^2)+year+
           origin+displacement:horsepower)
summary(lm.fit6)
```

```r
# Not much change in the residuals - still see evidence of a non-linear
# relationships with some of the predictors and response. Normal Q-Q indicates
# a deviation from normal distribution and has a more accentuated curve
par(mfrow=c(2,2))
plot(lm.fit6)
```

```r
# Looking back at the plotted correlation matrix, it seems that weight,
# displacement, and horsepower all have a log relationship with mpg.
# I'll convert mpg to log and remove the other modifications I made to
# the predictors and see what happens.
# The F statistic jumped up quite a bit and has a p-value close to 0 and R2
# is now 89.17, which also improved. 
lm.fit7=lm(log(mpg)~displacement+horsepower+weight+year+origin+
           displacement:horsepower)
summary(lm.fit7)
```

```r
# Checking the residuals, there is much less of an observable pattern. There 
# seems to be more variance and a larger effect from the observation with high
# leverage, but overall, this model seems to be a better fit than our previous
# model
par(mfrow=c(2,2))
plot(lm.fit7)
```
